#### Simple Linear Regression (SLR)

This is the simplest form of Linear Regression.

Here we have a single input variable(predictor) for the output variable (target)

The input or predictor variable is the variable that helps predict the value of the output variable. It is commonly referred to as X.

The output or target variable is the variable that we want to predict. It is commonly referred to as y.



B0 - point where regression line crosses y axis

B1 - determines the slope of the estimated regression line

#### Difference between correlation and regression
Reference: https://www.graphpad.com/support/faq/what-is-the-difference-between-correlation-and-linear-regression/#:~:text=Simple%20linear%20regression%20relates%20X,form%20Y%20%3D%20a%20%2B%20bX.&text=Both%20quantify%20the%20direction%20and,(b)%20will%20be%20negative.


Correlation quantifies the direction and strength of the relationship between two numeric variables, X and Y, and always lies between -1.0 and 1.0.

Simple linear regression relates X to Y through an equation of the form Y = a + bX.



Key Similarities:

-Both quantify the direction and strength of the relationship between two numeric variables.

-When the correlation (r) is negative, the regression slope (b) will be negative. 

-When the correlation is positive, the regression slope will be positive. 

-The correlation squared (r2 or R2) has special meaning in simple linear regression. It represents the proportion of variation in Y explained by X.
